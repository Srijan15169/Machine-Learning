```python
#maniputate the value of global variable from the function

x = 7
y = 9

def new_func():
  global x
  global y

  x = 3
  y = 4
  z = (x + y)
  print(z)
```

```python
print(x + y)

new_func()

print(x + y)
```

```python
#Classes in Python

class people:

  people1 = 'John'
  people2 = 'Bipul'

  def fun(self,x):
    print('Hello',self.people1,'and',x)

  def fun2(self):
    print('Welcome',self.people2)
```

```python
#Classes - constructor

class people:

  count = 0

  def __init__(self,name):     #Constructor
    self.n = name
    people.count += 1

  def say_hello(self):
    print('Hello, I am' , self.n)

  def total_count(self):
    print(self.count)
```

```python
p1 = people('Bipul')
p1.say_hello()
p1.total_count()
```

```python
p2 = people('John')
p2.say_hello()
p2.total_count()
```

```python
p3 = people('Joy')
p3.say_hello()
p3.total_count()
```

```python
p4 = people('Amit')
p4.say_hello()
p4.total_count()
```
