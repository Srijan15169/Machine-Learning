```python
import numpy as nmp  
import matplotlib.pyplot as mtplt 
import pandas as pd 
```

```python
def estimate_coeff(p, q):   
    n1 = nmp.size(p)
    m_p = nmp.mean(p)  
    m_q = nmp.mean(q)  
    SS_pq = nmp.sum(q * p) - n1 * m_q * m_p  
    SS_pp = nmp.sum(p * p) - n1 * m_p * m_p  
    b_1 = SS_pq / SS_pp  
    b_0 = m_q - b_1 * m_p  
    return (b_0, b_1)
```

```python
def plot_regression_line(p, q, b):  
    mtplt.scatter(p, q, color = "m", s = 30)  
    q_pred = b[0] + b[1] * p  
    mtplt.plot(p, q_pred, color = "g")  
    mtplt.xlabel('p')  
    mtplt.ylabel('q')  
    mtplt.show() 
```

```python
def main():  
    data = pd.read_csv("Salary_Data.csv")
    p = data["YearsExperience"] 
    q = data["Salary"]
    b = estimate_coeff(p, q)  
    print("Estimated coefficients are : ".format(b[0], b[1]))  
    plot_regression_line(p, q, b)  
  
if __name__ == "__main__":  
    main()  
```
